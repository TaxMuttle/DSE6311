```python
# Extract the year from the 'observation_date' column
df_medin['Year'] = pd.to_datetime(df_medin['observation_date']).dt.year

# Melt the DataFrame to long format
# The first column is 'observation_date' and the last is 'Year', so we'll use all columns in between as value variables
cols_to_melt = df_medin.columns[1:-1]
df_melted = df_medin.melt(id_vars=['Year'], value_vars=cols_to_melt, var_name='State_Code', value_name='Income')

# Extract state abbreviation from the State_Code column
df_melted['State'] = df_melted['State_Code'].apply(lambda x: x.split('MEHOINUS')[-1][:2])

# Select and reorder the desired columns
df_medin = df_melted[['Year', 'State', 'Income']]

# Display the resulting DataFrame
df_medin
```

```python
df_medin = df_medin[df_medin['Year'] >= 2000]
df_medin = df_medin[df_medin['Year'] <= 2019]
```

```python
df_medin = df_medin.sort_values(by=['Year', 'State'])
```
