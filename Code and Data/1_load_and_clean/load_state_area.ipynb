```python
df_area
```

```python
df_area = df_area[['Area', 'Total_SqMi']]
```

```python
df_area = df_area[df_area['Area'] != "Total"]
df_area = df_area[df_area['Area'] != "United States"]
df_area = df_area[df_area['Area'] != "District of Columbia"]
df_area = df_area[df_area['Area'] != "NaN"]
df_area = df_area[df_area['Area'] != "Puerto Rico"]
df_area = df_area[df_area['Area'] != "Island Areas"]
df_area = df_area[df_area['Area'] != "American Samoa"]
df_area = df_area[df_area['Area'] != "Guam"]
df_area = df_area[df_area['Area'] != "Northern Mariana Islands"]
df_area = df_area[df_area['Area'] != "U.S. Virgin Islands"]
```

```python
# Create a dictionary to map full state names to abbreviations
state_abbreviations = {
    'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR', 'California': 'CA',
    'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE', 'Florida': 'FL', 'Georgia': 'GA',
    'Hawaii': 'HI', 'Idaho': 'ID', 'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA',
    'Kansas': 'KS', 'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD',
    'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS', 'Missouri': 'MO',
    'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV', 'New Hampshire': 'NH', 'New Jersey': 'NJ',
    'New Mexico': 'NM', 'New York': 'NY', 'North Carolina': 'NC', 'North Dakota': 'ND', 'Ohio': 'OH',
    'Oklahoma': 'OK', 'Oregon': 'OR', 'Pennsylvania': 'PA', 'Rhode Island': 'RI', 'South Carolina': 'SC',
    'South Dakota': 'SD', 'Tennessee': 'TN', 'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT',
    'Virginia': 'VA', 'Washington': 'WA', 'West Virginia': 'WV', 'Wisconsin': 'WI', 'Wyoming': 'WY'
}

# Add a new column 'Abbreviation' to df_area based on the 'Area' column
df_area['State'] = df_area['Area'].map(state_abbreviations)

# Display the updated DataFrame
df_area
```

```python
df_area = df_area.dropna(subset=['Area'])
display(df_area.head())
```

```python
df_area = df_area.reset_index(drop=True)
display(df_area.head())
```

```python
df_area.sort_values(by='State', inplace=True)
```

```python
#mb moved import to top

# Create a list of years from 2000 to 2019
years = list(range(2000, 2020))

# Get the unique states and their Total_SqMi from df_area
state_area_data = df_area[['State', 'Total_SqMi']].drop_duplicates()

# Create a new DataFrame with all combinations of years and states
df_years = pd.DataFrame({'Year': years})

# Merge the state_area_data with the years DataFrame to get all year-state combinations
df_area_expanded = pd.merge(df_years, state_area_data, how='cross')

# Sort the DataFrame by State and Year for better readability
df_area_expanded.sort_values(by=['State', 'Year'], inplace=True)

# Reset the index
df_area_expanded.reset_index(drop=True, inplace=True)

# Assign the new expanded DataFrame back to df_area
df_area = df_area_expanded

# Display the head and tail of the updated DataFrame
display(df_area.head())
display(df_area.tail())
```

```python
df_area = df_area.sort_values(by=['Year', 'State'])
```

```python
df_area = df_area.reset_index(drop=True)
```
