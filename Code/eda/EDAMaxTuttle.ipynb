```python
# prompt: I want the current column values that are years to be put into their own year column and I want the current MSN (that repeat for each state) to be the columns

import pandas as pd
# The `temp3` DataFrame has columns 'State', 'MSN', and year columns like '1960', '1961', etc.
# We need to unpivot the year columns into rows and then pivot the 'MSN' column into columns.

# Step 1: Unpivot the year columns.
# We use pd.melt to convert the year columns (columns 2 onwards in temp3) into rows.
# 'State' and 'MSN' will be used as identifiers, 'variable_name' will hold the year, and 'value_name' will hold the energy value.
df_melted = pd.melt(temp3, id_vars=['State', 'MSN'], var_name='Year', value_name='Energy_Value')

# Step 2: Pivot the 'MSN' column into new columns.
# We use pd.pivot_table to restructure the DataFrame.
# 'State' and 'Year' will be the new index (rows), 'MSN' will be the columns, and 'Energy_Value' will be the values.
# We use aggfunc='first' because for each State, Year, and MSN combination, there should only be one Energy_Value.
df_pivot = pd.pivot_table(df_melted, values='Energy_Value', index=['State', 'Year'], columns='MSN', aggfunc='first')

# Reset the index to turn 'State' and 'Year' back into columns.
df_pivot = df_pivot.reset_index()

# Print the resulting DataFrame
print(df_pivot.head())
df_pivot.columns
df_pivot
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the median income data
median_income = pd.read_csv("annual.csv")

# Extract the year from the 'observation_date' column
median_income['Year'] = pd.to_datetime(median_income['observation_date']).dt.year

# Melt the DataFrame to long format
# The first column is 'observation_date' and the last is 'Year', so we'll use all columns in between as value variables
cols_to_melt = [col for col in median_income.columns if 'MEHOINUS' in col]
df_melted = median_income.melt(id_vars=['Year'], value_vars=cols_to_melt, var_name='State_Code', value_name='Income')

# Extract state abbreviation from the State_Code column
df_melted['State'] = df_melted['State_Code'].apply(lambda x: x.split('MEHOINUS')[-1][:2])

# Select and reorder the desired columns
median_income_long = df_melted[['Year', 'State', 'Income']]

# Select data for the latest year
latest_year = median_income_long['Year'].max()
median_income_latest_year = median_income_long[median_income_long['Year'] == latest_year]

# Sort the states by median income for the latest year
median_income_latest_year = median_income_latest_year.sort_values(by='Income', ascending=False)

# Create a bar plot of median income by state for the latest year
plt.figure(figsize=(14, 8))
sns.barplot(data=median_income_latest_year, x='State', y='Income', hue='State', palette='viridis', legend=False)
plt.title(f'Median Income by State in {latest_year}')
plt.xlabel('State')
plt.ylabel('Median Income')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Merge the median income and energy generation datasets
merged_df = pd.merge(median_income_long, energy_generation, left_on=['Year', 'State'], right_on=['YEAR', 'STATE'], how='inner')

# Drop the duplicate year and state columns from the energy_generation DataFrame
merged_df = merged_df.drop(columns=['YEAR', 'STATE'])

# Display the head of the merged DataFrame
display(merged_df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Filter the DataFrame to include only rows where 'Energy Source' is 'Total'
total_energy_df = merged_df[merged_df['ENERGY SOURCE'] == 'Total'].copy()

# Create a scatter plot of median income vs total energy generation
plt.figure(figsize=(10, 6))
sns.scatterplot(data=total_energy_df, x='Income', y='GENERATION (Megawatthours)', alpha=0.5)
plt.title('Median Income vs. Total Energy Generation')
plt.xlabel('Median Income')
plt.ylabel('Total Energy Generation (Megawatthours)')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Calculate income quartiles
# Ensure the 'Income' column is numeric, coercing errors will turn non-numeric into NaN
merged_df['Income'] = pd.to_numeric(merged_df['Income'], errors='coerce')

# Drop rows with NaN in 'Income' after coercion
merged_df.dropna(subset=['Income'], inplace=True)

merged_df['Income_Quartile'] = pd.qcut(merged_df['Income'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

# Group by income quartile and calculate the mean energy generation
energy_by_quartile = merged_df.groupby('Income_Quartile')['GENERATION (Megawatthours)'].mean().reset_index()

# Create a bar plot of energy generation by income quartile
plt.figure(figsize=(10, 6))
sns.barplot(data=energy_by_quartile, x='Income_Quartile', y='GENERATION (Megawatthours)', hue='Income_Quartile', palette='viridis', legend=False)
plt.title('Average Energy Generation by Income Quartile')
plt.xlabel('Income Quartile')
plt.ylabel('Average Energy Generation (Megawatthours)')
plt.tight_layout()
plt.show()
```
